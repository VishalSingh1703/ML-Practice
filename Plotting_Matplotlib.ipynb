```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from imageio import imread
```

```python
plt.plot(3, 4, 'ro', label = 'red circle')
plt.plot(2, 3, 'b^', label = 'blue triangle')
plt.legend(['red', 'blue'])
plt.legend()
# plt.show()
```

```python
x = [1, 2, 4]
y = [3, 4, -1]
plt.plot(x, y, 'ms--')
```

```python
x = np.linspace(1, 3*np.pi, 101)
y = np.sin(x)

plt.plot(x, y, 'm:')
```

```python
plt.plot([0, 1], [0, 1], 'm', linewidth = 14)
plt.plot([0, 1], [1, 0], 'm')
plt.axis('square')
```

```python
plt.subplot(1, 2, 1)
plt.plot(np.random.randn(10))

plt.subplot(2, 2, 2)
plt.plot(np.random.randn(10))
```

```python
fig,ax = plt.subplots(1, 2, figsize = (10, 4))
x = np.arange(20)
ax[0].plot(x, x**2, 'k')
ax[0].plot(x, x, 'r')
ax[1].plot(x, np.sqrt(x), 'g')
plt.show()
```

```python
fig, ax = plt.subplots(2, 2)
ax[0, 0].plot(np.random.randn(4, 4))
ax[0, 1].plot(np.random.randn(4, 4))
ax[1, 0].plot(np.random.randn(4, 4))
ax[1, 1].plot(np.random.randn(4, 4))
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(2, 2)
for i in range(0, 2):
  for j in range(0, 2):
    ax[i, j].plot(np.random.randn(4, 4))

plt.tight_layout()
plt.show()
```

```python
M = np.random.randint(10, size = (4, 4))
# print(M), print(" ")
# print(M.flatten())

fig, ax = plt.subplots(3, 3)
for i in ax.flatten():
  i.plot(np.random.randn(3, 4))
plt.show()
```

```python
x = np.linspace(-3, 3, 101)
plt.plot(x, x, label = 'y = x')
plt.plot(x, x**2, label = 'y = x**2')
plt.plot(x, x**3, label = 'y = x**3')
plt.legend()
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('An awesome plot')
plt.show()
```

```python
fig, ax = plt.subplots()
x = np.linspace(0, 10, 100)

for i in np.linspace(0, 1, 50):
  ax.plot(x, x**i, color = [i/2, 0, i])

plt.show()
```

```python
n = 100
D = np.zeros((n, 2))
D[:,0] = np.linspace(0, 10, n) + np.random.randn(n)
D[:,1] = D[:,0]**2 + np.random.randn(n)*10

# plt.plot(D[:,0], D[:,1], 'o')
sns.jointplot(x = D[:,0], y = D[:,1])
plt.show()
```

```python
df = pd.DataFrame(data = D, columns = ['var1', 'var2'])
sns.jointplot(x='var1', y='var2', data=df, kind = 'kde', color = 'purple')
df
```

```python
n=200
x = np.linspace(-1, 1, n)
y1 = x**2
y2 = np.sin(x*3)
y3 = np.exp(-10*x**2)
sns.scatterplot(x = y1, y = y2, hue = y3, legend = False)
plt.show()
```

```python
sns.regplot(x = df.columns[0], y = df.columns[1], data = df)
plt.title(f'Regression of {df.columns[0]} on {df.columns[1]}')
plt.show()
```

```python
a = 3
b = 5

M = np.random.randint(10, size = (a, b))
print(M)
# plt.plot(M, 'm-')
plt.imshow(M)
plt.colorbar()

for i in range(a):
  for j in range(b):
    plt.text(j, i, str(M[i,j]))
plt.show()
```
