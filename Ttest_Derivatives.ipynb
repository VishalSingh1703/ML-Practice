```python
import numpy as np
import matplotlib.pyplot as plt

import scipy.stats as stats # used for Ttest

import sympy as sym # used for derivatives
import sympy.plotting.plot as symplot

from IPython.display import display # for making equations look nicer
```

```python
n1 = 30 # samples in dataset 1
n2 = 40
mul1 = 1  # population mean in dataset 1
mul2 = 2

# generate data
data1 = mul1 + np.random.randn(n1)
data2 = mul2 + np.random.randn(n2)


plt.plot(np.zeros(n1), data1, 'ro', markerfacecolor = 'w', markersize = 14)
plt.plot(np.ones(n2), data2, 'bs', markerfacecolor = 'w', markersize = 14)
plt.xlim([-1, 2])
plt.xticks([0, 1], labels = ['Group1', 'Group2'])
plt.show()
```

```python
t, p = stats.ttest_ind(data1, data2)
print(t)
print(p)
```

```python
fig = plt.figure(figsize = (10, 4))
plt.rcParams.update({'font.size':12}) #change font size

plt.plot(0 + np.random.randn(n1)/15, data1, 'ro', markerfacecolor = 'w', markersize = 14)
plt.plot(1 + np.random.randn(n2)/15, data2, 'bs', markerfacecolor = 'w', markersize = 14)
plt.xlim([-1, 2])
plt.xticks([0, 1], labels = ['Group 1', 'Group 2'])
plt.title(f't = {t: 2f}, p = {p: 3f}')

plt.show()
```

```python
x = sym.symbols('x') # Create symbolic variables in sympy

fx = 2*x**2 # Create a function

df = sym.diff(fx, x) # Compute its derivative

print(fx)
print(df)

# plotting them
symplot(fx, (x, -4, 4), title = 'The Function')
plt.show()

symplot(df, (x, -4, 4), title = 'The Derivative')
plt.show()
```

```python
# repeat with relu and sigmoid
# create symbolic functions
relu = sym.Max(0, x)
sigmoid = 1 / ( 1 + sym.exp(-x) )

#graph the functions
p = symplot(relu, (x, -4, 4), label = 'ReLU', show = False, line_color = 'blue')
p.extend( symplot(sigmoid, (x, -4, 4), label = 'Sigmoid', show = False, line_color = 'red') )
p.legend = True
p.title = 'The Functions'
p.show()

#graph the derivatives
p = symplot(sym.diff(relu), (x, -4, 4), label = 'df(ReLU)', show = False, line_color = 'blue')
p.extend( symplot(sym.diff(sigmoid), (x, -4, 4), label = 'df(Sigmoid)', show = False, line_color = 'red') )
p.legend = True
p.title = 'The Derivates'
p.show()
```

```python
x = sym.symbols('x') # create symbolic variables in sympy

# creating two functions
fx = 2*x**2
gx = 4*x**3 - 3*x**4

# compute individual derivatives
df = sym.diff(fx)
dg = sym.diff(gx)

# apply the product rule "manually"
manual = df*gx + fx*dg

# via sympy
viasympy = sym.diff(fx*gx)

print('The functions: ')
display(fx)
display(gx)
print(' ')

display(manual)
display(viasympy)
```

```python
# repeat with chain rule

fx = ( x**2 + 4*x**3 )**5

print('The Functions')
display(fx)
print(' ')

print('The Derivatives')
display(sym.diff(fx))
```
